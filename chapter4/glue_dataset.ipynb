## GLUE 데이터셋

NLP 모델을 평가하기 위한 데이터셋 중에서 가장 많이 사용되는 데이터셋

GLUE : General Language Understanding Evaluation 의 약어

GLUE 는 9개의 서브태스크로 언어 모델의 성능을 평가함

9개의 서브태스크는
- CoLA, SST-2, MRPC, QQP, STS-B, MNLI, QNLI, RTE, WNLI

### CoLA

Corpus of Linguistic Acceptability

영어 문장의 문법적인 수용성을 평가하는 데이터셋

- 0: 문법적으로 수용되지 않는 문장
- 1: 문법적으로 수용되는 문장

Unbalanced Classification Data

Matthew Correlation Coefficient를 사용해서 평가

(Example)
```
# label=0
They drank the pub.
The professor taked us.

# label=1
They drank the pub only.
The professor talked us into a stupor.
We yelled ourselves hoarse
```

### SST-2

Stanford Sentiment Treebank

영화 리뷰에 대한 사람의 감정을 positive/negative 로 나눈 데이터셋

- 0: nagative
- 1: positive

accuracy 를 통해 평가

### MRPC

Microsoft Research Paraphrase Corpus

마이크로소프트에서 공개한 문장 쌍 데이터셋으로 같은지 다른지를 0과 1로 판단

- 0: not_same
- 1: same

accuracy 와 F1 스코어로 측정

### QQP

Quora 로부터 확보한 질문을 쌍으로 연결해 둔 데이터셋

Quora 는 네이버 지식인과 비슷한 질의응답 웹사이트. 중복인지 아닌지 구분한 데이터셋.

- 0: not same questions
- 1: same questions

accuracy 와 F1 스코어로 측정

### STS-B

Semantic Textual Similarity Benchmark

뉴스, 비디오, 이미지, 캡션 등으로부터 추출한 문장 쌍 데이터셋

각 데이터셋은 유사도 스코어를 1~5 사이의 값으로 구분한 문장 쌍들로 구성됨

Person and Spearman Correlation Coefficient 로 평가 측정

## MNLI

Multi-Genre Natural Language inference

가설(hypothesis)이 전제(premise)를 수반(entailment)하는지, 모순(contradiction)하는지, 중립(neutral)적인지를 나타내는 데이터셋

accuracy 로 측정

### QNLI

스탠퍼드 질문/응답 데이터셋으로부터 생성한 데이터셋

문장과 질문을 쌍으로 두고 질문에 대한 응답이 문장 내에 있는지 없는지를 판단하는 데이터셋 (not_entailment, entailment)

accuracy 로 측정

### RTE

Recognizing Textual Entailment

RTE1, RTE2, RTE3, RTE5 를 합친 데이터셋

뉴스와 위키피디아 텍스트로 이뤄져있음

두 문장을 쌍으로 만들어서 그 두 문장이 서로 수반되는 문장인지, 그렇지 않은지를 분류한 데이터셋 (not_entailment, entailment)

accuracy 로 측정

### WNLI

Winograd Schema Challenge 라는 독해 능력 태스크로부터 생성한 데이터셋

Winograd Schema Challenge 는 문장에 특정 단어를 가리키는 레퍼런스를 선택지로부터 찾는 문제

accuracy 로 측정
